```python
import copy
df1 = copy.deepcopy(df)
df1.head()
```

```python
df1.drop(["RowNumber", "CustomerId", "Surname"], axis = 1, inplace = True)
```

```python
# Converting Categorical variables
Location = pd.get_dummies(df1["Geography"], drop_first = True)
sex = pd.get_dummies(df1["Gender"], drop_first = True)
```

```python
df1 = pd.concat([df1,Location,sex], axis=1)
df1.drop(["Geography","Gender"], axis=1, inplace = True)
```

```python
df1.head()
```

```python
df1.info()
```

```python
x = df1.drop("Exited", axis = 1)
y = df1["Exited"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.8, random_state = 42)
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
log = LogisticRegression()
log.fit(x_train, y_train)
pred = log.predict(x_test)
accuracy_score(pred, y_test)
print("Accuracy score : \n", accuracy_score(pred, y_test))
print("Classification report : \n", classification_report(pred, y_test))
print("Confusion matrix : \n", confusion_matrix(pred, y_test))
```

```python
# Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 30)
rf.fit(x_train, y_train)
rf_pred = rf.predict(x_test)
print("Accuracy score : \n", accuracy_score(rf_pred, y_test))
print("Classification report : \n", classification_report(rf_pred, y_test))
print("Confusion matrix : \n", confusion_matrix(rf_pred, y_test))
```

```python
# Grid Search Cv
from sklearn.model_selection import GridSearchCV
param_grid = [{'n_estimators': [3, 10, 30], 'max_features': [2, 4, 6, 8, 10, 12]},{'bootstrap': [False], 'n_estimators': [3, 10], 'max_features': [2, 3, 4]}]
forest_reg = RandomForestClassifier()
grid_search = GridSearchCV(forest_reg, param_grid, cv=5,
 scoring='neg_mean_squared_error')
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_params_
```

```python
grid_rf = RandomForestClassifier(n_estimators=30,max_features=4)
grid_rf.fit(x_train,y_train)
grid_rf_pred = grid_rf.predict(x_test)
```

```python
print("Accuracy score : ", accuracy_score(grid_rf_pred,y_test))
```
